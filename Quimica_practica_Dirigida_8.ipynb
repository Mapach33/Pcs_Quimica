```python
data = pd.read_csv("graficaA.csv")
# Crea un gráfico de dispersión para mostrar la relación entre P y V
plt.figure(figsize=(6, 4))
plt.scatter(data['x'], data['y'], marker='o', color='b')
plt.xlabel('Volumen (L)')
plt.ylabel('Presión (atm)')
plt.title('Ley de Boyle-Mariotte')
plt.grid(True)

plt.show()
```

```python
link = 'https://es.wikipedia.org/wiki/Ley_de_Boyle-Mariotte'
tabla = pd.read_html(link)

data = tabla[2]
plt.figure(figsize=(6, 4))
plt.scatter(data['V (L)'], data['P (atm)'], marker='o', color='b')
plt.xlabel('Volumen (L)')
plt.ylabel('Presión (Pa)')
plt.title('Ley de Boyle-Mariotte')
plt.grid(True)

plt.plot(data['V (L)'], data['P (atm)'], color='r', linestyle='--', label='Ley de Boyle-Mariotte')

plt.legend()

plt.show()
```
